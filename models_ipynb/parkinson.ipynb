```python
df = df.drop(['name'], axis=1)
```

```python
df['status'].value_counts()
```

```python
features = [ 'MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
       'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP',
       'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'MDVP:APQ', 'Shimmer:DDA', 'NHR', 'HNR', 'RPDE', 'DFA',
       'spread1', 'spread2', 'D2', 'PPE']
```

```python
df['status'].unique()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(df[features].corr(),annot = True)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, train_size = 0.80)
```

```python
X_train = train[features]
Y_train = train['status']

X_test = test[features]
Y_test = test['status']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
%matplotlib inline
```

```python
model = RandomForestClassifier(n_estimators = 500)
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
Y_pred = model.predict(X_test)
print(Y_pred)
```

```python
cm = confusion_matrix(Y_test, Y_pred)
cm
```

```python
plt.figure(figsize =(10,7))
sns.heatmap(cm, annot = True)
plt.xlabel('Predicted')
plt.ylabel('real')
```

```python
score = accuracy_score(Y_test, Y_pred)*100
acc = round(score, 2)
inac = round(100-acc,2)
corr = (score*len(test))/100
print('Accuracy: ',acc, '%, ', corr, 'correct detections')
print('Inaccuracy: ',inac, '%, ', len(test)-corr, 'Incorrect detections')
```
